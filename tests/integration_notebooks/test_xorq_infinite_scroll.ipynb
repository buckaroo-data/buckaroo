```python
import buckaroo
import xorq.api as xo
from buckaroo.xorq_buckaroo import XorqBuckarooInfiniteWidget

# Create a large predictable xorq expression for testing infinite scroll
# Row i has: int_col = i + 10, str_col = "foo_{i + 10}"
# 5000 rows ensures we need multiple data fetches to satisfy scroll requests.
N_ROWS = 5000
expr = xo.memtable({
    "row_num": list(range(N_ROWS)),
    "int_col": [i + 10 for i in range(N_ROWS)],
    "str_col": [f"foo_{i + 10}" for i in range(N_ROWS)],
})
print(f"✅ Created xorq expression with {N_ROWS} rows")
print(f"   First row: row_num=0, int_col=10, str_col='foo_10'")
print(f"   Last row: row_num={N_ROWS-1}, int_col={N_ROWS-1+10}, str_col='foo_{N_ROWS-1+10}'")

widget = XorqBuckarooInfiniteWidget(expr, record_transcript=True)
print("✅ XorqBuckarooInfiniteWidget created with transcript recording enabled")
widget
```
