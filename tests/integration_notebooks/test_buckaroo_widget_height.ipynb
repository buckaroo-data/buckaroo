# Buckaroo widget — height regression notebook

Two cells: a tiny dataframe (4 rows) and a larger one (200 rows). The
Playwright spec `pw-tests/jupyter-buckaroo-height.spec.ts` opens this
notebook at varying viewport heights and asserts:

  - The widget renders ag-grid cells in both output areas.
  - The grid bottom never sits more than a few pixels above the last
    data row (no dead band inside the grid).
  - The grid hugs its content for the small DF — Jupyter notebooks
    naturally want auto-sized embeds because their output container
    grows with the cell content.

```python
import buckaroo  # noqa
import pandas as pd
from buckaroo import BuckarooWidget

small_df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Dianne'],
    'age': [25, 30, 35, 28],
    'score': [85.5, 92.0, 78.3, 88.1],
})
BuckarooWidget(small_df)
```

```python
large_df = pd.DataFrame({
    'idx': list(range(200)),
    'val': [i * 1.5 for i in range(200)],
    'lbl': [f'row_{i}' for i in range(200)],
})
BuckarooWidget(large_df)
```
