```python
import pandas as pd

# Install and import buckaroo
import sys
if "pyodide" in sys.modules:
    import micropip
    # In JupyterLite, we need a full URL to the wheel file
    # Use js.location to get the current origin
    import js
    await micropip.install("traitlets", deps=False)
    await micropip.install("ipywidgets", deps=False)
    await micropip.install("psygnal", deps=False)
    await micropip.install("anywidget", deps=False)
    await micropip.install("typing_extensions", deps=False)
    base_url = js.location.origin
    
    # Check Pyodide version
    try:
        import pyodide
        print(f"✅ Pyodide version: {pyodide.__version__}")
    except Exception as e:
        print(f"⚠️  Could not get Pyodide version: {type(e).__name__}: {e}")
    
    # Check Python version
    import sys
    print(f"✅ Python version: {sys.version}")
    
    # Check if pyarrow was preloaded via loadPyodideOptions
    try:
        import pyarrow as pa
        print(f"✅ pyarrow preloaded and available: {pa.__version__}")
    except ImportError as import_err:
        print(f"⚠️  pyarrow not available: {type(import_err).__name__}: {import_err}")
        print("⚠️  pyarrow should have been preloaded via loadPyodideOptions config")
    
    # Check if fastparquet was preloaded
    try:
        import fastparquet
        print(f"✅ fastparquet preloaded and available: {fastparquet.__version__}")
    except ImportError as import_err:
        print(f"⚠️  fastparquet not available: {type(import_err).__name__}: {import_err}")
        print("⚠️  fastparquet should have been preloaded via loadPyodideOptions config")
    
    # Install buckaroo wheel
    wheel_url = f"{base_url}/files/buckaroo-0.12.6-py3-none-any.whl"
    await micropip.install(wheel_url, deps=True)

import buckaroo
```
