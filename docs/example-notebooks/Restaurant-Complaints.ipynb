# Restaurant Complaints

Open `/Users/paddy/buckaroo/restaurant-complaints.parquet`, reorder columns as `[businessname, comments, violation, violdesc, ...rest]`, give `comments` a long `max_length`, and use a 5-line-tall row height with `wrapText` so long comments wrap inside the cell.

Note: the column in the parquet is `violdesc`, not `violationdesc`.

```python
import polars as pl
from buckaroo.polars_buckaroo import PolarsBuckarooInfiniteWidget

df = pl.read_parquet('/Users/paddy/buckaroo/restaurant-complaints.parquet')

front = ['businessname', 'comments', 'violation', 'violdesc']
df = df.select(front + [c for c in df.columns if c not in front])

extra_grid_config = {'rowHeight': 105, 'pinnedRowHeight': 21}  # ~5 lines at the default ~21px line height
column_config_overrides = {
    'comments': {
        'displayer_args': {'displayer': 'string', 'max_length': 5000,
                           # 'highlight_color':'red',
                           # 'highlight_phrase':['area'],
                          },
        'ag_grid_specs': {'wrapText': True, 'width':400, 'maxWidth':800},
        'delete_keys': ['tooltip_config'],
    },
}
```

```python
PolarsBuckarooInfiniteWidget(df, init_sd=column_config_overrides, extra_grid_config=extra_grid_config,
                              component_config = {'dfvHeight': 900})
```

```python
from buckaroo.buckaroo_widget import BuckarooWidget

BuckarooWidget(
    df,
    column_config_overrides={
        'comments': {
            'displayer_args': {'displayer': 'string', 'max_length': 2000},
            'ag_grid_specs': {'wrapText': True},
        },
    },
    extra_grid_config={'rowHeight': 105},  # ~5 lines at the default ~21px line height
)
```
