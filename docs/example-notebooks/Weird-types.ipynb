## Individual type checks

```python
# Polars Duration — was showing blank cells
pl.DataFrame({
    'short': pl.Series([100, 500, 1000], dtype=pl.Duration('us')),
    'medium': pl.Series([60_000_000, 3_600_000_000, 86_400_000_000], dtype=pl.Duration('us')),
    'long': pl.Series([86_400_000_000 * 365, 86_400_000_000 * 30, 86_400_000_000 * 7], dtype=pl.Duration('us')),
})
```

```python
# Polars Time — was showing blank cells
pl.DataFrame({'time': [dt.time(0, 0), dt.time(12, 30, 45), dt.time(23, 59, 59)]})
```

```python
# Polars Decimal — was losing precision (showed 100 instead of 100.50)
pl.DataFrame({'price': pl.Series(['9.99', '100.50', '1234.00']).cast(pl.Decimal(10, 2))})
```

```python
# Polars Categorical
pl.DataFrame({'color': pl.Series(['red', 'green', 'blue', 'red']).cast(pl.Categorical)})
```

```python
# Pandas Period
pd.DataFrame({'month': pd.period_range('2024-01', periods=6, freq='M')})
```

```python
# Pandas Interval
pd.DataFrame({'range': pd.arrays.IntervalArray.from_breaks(range(6))})
```

```python
# Pandas Timedelta
pd.DataFrame({'elapsed': pd.to_timedelta(['1 days 02:03:04', '0 days 00:00:01', '365 days', '0:00:00.001', '0:00:00.000100'])})
```
